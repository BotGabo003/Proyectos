```python
#Objetivo: Crear una clase que calcule el área y perímetro de un rectángulo.

class Calculo:
    def __init__(self,largo,ancho):
        self.largo=largo 
        self.ancho=ancho
    def area(self):
        return self.largo*self.ancho
    def perimetro(self):
        return (2*self.largo)+(2*self.ancho)

imprimir=Calculo(4,5)
print (f"El rectangulo da como Perimetro {imprimir.perimetro()} y el Area es {imprimir.area()}")

      
```

```python
#Cuenta de banco
class Cuenta:
     
    def __init__(self,saldo):
        self.saldo=saldo
    def deposito(self,plata):
        self.saldo+=plata
        return self.saldo
    def retiro(self,plata):
        self.saldo-=plata
        return self.saldo
    accion=str(input("Usted desea depositar o retirar dinero"))
def pregunta(accion):
    if accion=="depositar":
        return True
    else:
        accion=="retirar"
        return False

saldo=30000
cuenta=Cuenta(saldo)
accion=str(input("Usted desea depositar o retirar dinero"))
accion2=pregunta(accion)
if  accion2 == True:
    plata=int(input("Ingrese el dinero"))
    imprimir=cuenta.deposito(plata)
    print(f"El saldo de la cuenta luego del deposito es {imprimir}")
else:
    plata=int(input("Ingrese el dinero"))
    imprimir=cuenta.retiro(plata)
    print(f"El saldo de la cuenta luego del retiro es {imprimir}") 



    
```
